```python
# Using the plot_animal() function defined earlier

import pandas as pd
import matplotlib.pyplot as plt

word_trends = pd.read_csv('animal-word-trends-menageries.csv')
word_trends
```

```python
# Plotting code - define it to a reusable plot_animal() function

def plot_word_trend( animal ):
    trend = word_trends.query('word == @animal')                    #access the animal variable using the @animal notation
    plt.plot(trend['year'], trend['frequency'], label=animal)
    plt.ylabel('Frequency Per Million')
```

```python
# Explore exotic animal words during the menagerie period

plot_word_trend('tiger')
plt.axvspan(1750, 1835, color='magenta', label='Menagerie Period')
plt.title('The menagerie period and word frequency for "tiger"')
```
