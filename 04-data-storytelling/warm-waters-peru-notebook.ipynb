```python
# Load data

import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', 10)

word_trends = pd.read_csv('animal-word-trends-peru.csv')
word_trends
```

```python
# Define plot_animal() function used earlier

def plot_animal( animal ):
	trend = word_trends.query("word == @animal")
	plt.plot(trend['year'], trend['frequency'], label=animal)
	plt.ylabel('Frequency Per Million')
```

```python
# Explore trends for fish words relative to El Niño years

plot_animal('sardine')
for year in [1965, 1973, 1983, 1987]:
    plt.axvline( year, label='El Niño Event', color='magenta', ls='--')

plt.xlim(1960, 1990)
plt.title('El Niñoevents and word frequency for "sardine"')
```
