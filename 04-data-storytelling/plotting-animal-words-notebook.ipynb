```python
# Load data

import pandas as pd
import matplotlib.pyplot as plt

word_trends = pd.read_csv('animal-word-trends-intro.csv')
word_trends
```

```python
# Use query() to view historical change in usage frequency for the word 'horse'

trend = word_trends.query('word == "horse"')
trend
```

```python
# Plot trend over time usig a line plot

plt.plot(trend['year'], trend['frequency'])
plt.ylabel('Frequency Per Million')
plt.title('Word Frequency For "Horse" Over Time')
```

```python
# Plotting code - define it to a reusable function

def plot_word_trend( animal ):
    trend = word_trends.query('word == @animal')                    #access the animal variable using the @animal notation
    plt.plot(trend['year'], trend['frequency'], label=animal)
    plt.ylabel('Frequency Per Million')
```

```python
# Re-plot for 'Horse' using the defined function

plot_word_trend('horse')
plt.axvline(1886, color='orange', ls='--')
plt.title('Word usage for "horse" and the year the automobile was invented')
```

```python
# Plot for 'penguin' & 'dinosaur' using the defined function

plot_word_trend('penguin')
plot_word_trend('dinosaur')
plt.legend()
```

```python
# Exploring and forming hypotheses for 'Goat'

plot_word_trend('goat')
plt.axvline(1869, color='magenta', ls='--')
plt.title('Transcontinental railway and word usage for "goat"')
```

```python
# Exploring and forming hypotheses for 'Lobster'

plot_word_trend('lobster')
plt.axvline(1869, color='magenta', ls='--')
plt.title('Transcontinental railway and word usage for "lobster"')
```

```python
# Exploring and forming hypotheses for 'Shrimp'

plot_word_trend('shrimp')
plt.axvline(1869, color='magenta', ls='--')
plt.title('Transcontinental railway and word usage for "shrimp"')
```

```python
# Exploring and forming hypotheses for 'Tuna'

plot_word_trend('tuna')
plt.axvline(1869, color='magenta', ls='--')
plt.title('Transcontinental railway and word usage for "tuna"')
```

```python
# Exploring and forming hypotheses for 'Goat, Lobster, Shrimp, Tuna'

plot_word_trend('goat')
plot_word_trend('lobster')
plot_word_trend('shrimp')
plot_word_trend('tuna')
plt.axvline(1869, color='magenta', ls='--')
plt.title('Transcontinental railway and word usage for "Goat, Lobster, Shrimp, Tuna"')
plt.legend()
```
