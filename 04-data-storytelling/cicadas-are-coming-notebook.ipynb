```python
# Load cicadas data

import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', 10)

word_trends = pd.read_csv('animal-word-trends-cicada.csv')
word_trends
```

```python
# Using the previous plot_animal() function

def plot_animal( animal ):
    trend = word_trends.query("word == @animal")
    plt.plot(trend['year'], trend['frequency'], label=animal)
    plt.ylabel('Frequency per million')
```

```python
# Plot animal word for 'cicada' between 1820 to 2020

plot_animal('cicada')
plt.xlim(1820, 2020)
```

```python
# Compare historical brood years with the word usage patterns between 1915 to 1970 (17-Years Broods)

plot_animal('cicada')
plt.xlim(1820, 2020)
plt.title('Brood X: Years Emerged')

for year in [1919, 1936, 1953]:
    plt.axvline(year, color='magenta', ls=':')
```
