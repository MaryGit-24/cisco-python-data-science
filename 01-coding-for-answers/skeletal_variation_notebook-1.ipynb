```python
# Load data
import pandas as pd
df = pd.read_csv('adult-human-skeleton.csv')
df
```

```python
# Count bones in each region
# value_count returns a series containing counts of unique values
df['region'].value_counts()
```

```python
# Calculate the proportion of hands and feet in the entire body
# Over 51% of bones are found in the hands and feet!
(54 + 52) / 206
```

```python
# Baby bones
# Sort values from column
# sort_values; sort by on either axis
df.sort_values(by='fused_from', ascending=False)
```

```python
# How many bones do infants have?
# Sum all fused from values
df['fused_from'].sum()
```

```python
# Human neck bones
# How many bones are in the human neck?
df.query('region == "neck"')
```

```python
# Bird neck bones
birds = pd.read_csv('bird-neck-bones.csv')
birds
```

```python
# Plot a bar chart of neck_vertebae using value_count
# Use sort_index to show the bird count ordered by number of neck vertebrae
bird_counts = birds['neck_vertebrae'].value_counts()
bird_counts = bird_counts.sort_index()
bird_counts.plot.bar()
```

```python
# Find the rows with the number of neck vertebrae
birds.query('neck_vertebrae == neck_vertebrae.max()')
```
